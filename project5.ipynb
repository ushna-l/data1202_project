```python
df = pd.read_csv('youtube_dataset.csv', encoding='ISO-8859-1')
```

```python
df = df.dropna(subset=['channeltype'])
```

```python
df.head(1000)
```

```python
channel_types = df['channeltype']
```

```python
print(channel_types.head(1000))
```

```python
def calculate_channel_distribution(df):
    top_1000 = df.head(1000)
    distribution = top_1000['channeltype'].value_counts()
    
    return distribution
```

```python
channel_distribution = calculate_channel_distribution(df)
print(channel_distribution)
```

```python
channel_distribution_df = channel_distribution.reset_index()
channel_distribution_df.columns = ["Channel Type", "Count"] 
```

```python
output_csv = "channel_distribution.csv"
channel_distribution_df.to_csv(output_csv, index=False, encoding="utf-8")
```
